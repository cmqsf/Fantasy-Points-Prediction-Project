```python
qb_rankings = pd.DataFrame(columns = ['QB', 'Points'])
for player in qbs['player']:
    ypred = getPredictedFantasyPoints(qbs, player, 155)
    pts = (ypred[0] + ypred[1])/2
    qb_rankings.loc[len(qb_rankings)] = [player, pts]
    
```
